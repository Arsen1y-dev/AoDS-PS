6.3. Модуль requests

Проверка системы

```python
from requests import get

print(get('http://127.0.0.1:5000').text)
```

Этот код выполняет простой HTTP GET-запрос к локальному серверу, работающему на порту 5000, и выводит текстовый ответ сервера.

get('http://127.0.0.1:5000'):
Выполняет HTTP GET-запрос к URL-адресу http://127.0.0.1:5000.
127.0.0.1 - это стандартный IP-адрес для обозначения локальной машины (вашего компьютера).
5000 - это номер порта, на котором работает сервер.
.text:
Извлекает текстовое содержимое ответа сервера. Ответ сервера может содержать различную информацию, такую как HTML-код страницы, JSON-данные или просто текстовое сообщение.
print(...):
Выводит полученный текст на консоль.
В итоге, этот код:
Отправляет запрос к локальному серверу.
Получает текстовый ответ от сервера.

Суммирование ответов

```python
from requests import get

address = 'http://' + input()
summ = 0
while data := int(get(address).text):
    summ += data
print(summ)
```

Этот код последовательно получает числа с сервера, расположенного по указанному пользователем адресу, и суммирует их до тех пор, пока сервер не вернёт 0.

address = 'http://' + input():
Спрашивает у пользователя адрес сервера (без http://) и добавляет префикс http:// к введённому адресу.
summ = 0:
Инициализирует переменную summ, в которой будет храниться сумма полученных чисел.
while data := int(get(address).text)::
Начинается цикл while, который выполняется, пока условие истинно.
Внутри условия используется оператор присваивания с помощью := (walrus operator).
get(address) выполняет HTTP GET-запрос к указанному адресу.
.text извлекает текстовое содержимое ответа сервера.
int(...) преобразует полученный текст в целое число.
Результат присваивается переменной data и одновременно используется как условие цикла.
Цикл продолжается, пока сервер возвращает ненулевые числа.
summ += data:
Внутри цикла полученное число data добавляется к переменной summ.
print(summ):
После завершения цикла (когда сервер вернул 0) выводится итоговая сумма на консоль.
В итоге, этот код:
Запрашивает у пользователя адрес сервера.
Повторяет следующие шаги:
Выполняет GET-запрос к серверу.
Преобразует текстовый ответ в число.
Добавляет полученное число к сумме.

Конкретное значение

```python
from requests import get

address = 'http://' + input()
key = input()
data = get(address).json()
print(data.get(key, 'No data'))
```

Этот код использует библиотеку requests для получения данных из API и извлечения определенного значения. 
2. Ввод адреса:
address = 'http://' + input()
Здесь происходит формирование адреса API. Сначала к строке 'http://' добавляется текст, введенный пользователем. Таким образом пользователь может указать адрес нужного API.
3. Ввод ключа:
key = input()
Здесь пользователь вводит ключ, по которому нужно извлечь значение из полученного JSON ответа.
4. Отправка запроса и получение данных:
data = get(address).json()
Функция get отправляет GET запрос на указанный адрес (address). Результат запроса (ответ сервера) преобразуется в JSON формат с помощью метода .json(). Полученный словарь сохраняется в переменной data.
5. Извлечение значения по ключу:
print(data.get(key, 'No data'))
С помощью метода .get(key, 'No data') из словаря data извлекается значение, соответствующее ключу key. Если ключ не найден, то выводится текст 'No data'.

Принцип работы:
Пользователь вводит адрес API и ключ.
Программа отправляет GET запрос на указанный адрес.
API возвращает данные в формате JSON.
Программа преобразует JSON в словарь Python.
Программа извлекает значение по указанному ключу из словаря.
Программа выводит найденное значение или сообщение об ошибке.

Суммирование ответов 3

```python
from requests import get
from sys import stdin

address = 'http://' + input()
ways = [i.strip() for i in stdin]
summ = 0
for way in ways:
    summ += sum(get(address + way).json())
print(summ)
```


Этот код также взаимодействует с API, но он обрабатывает несколько запросов, используя список путей, полученных из стандартного потока ввода (stdin).
2. Ввод адреса сервера:
address = 'http://' + input()
Считывает адрес сервера от пользователя и добавляет префикс http://.
3. Получение списка путей:
ways = [i.strip() for i in stdin]
Считывает строки из стандартного потока ввода (stdin).
Для каждой строки выполняет .strip(), чтобы удалить пробелы в начале и конце.
Создает список ways, содержащий полученные строки (пути).
4. Суммирование данных по каждому пути:
summ = 0
for way in ways:
    summ += sum(get(address + way).json())
Инициализирует переменную summ для хранения итоговой суммы.
Цикл for перебирает каждый путь way из списка ways.
Для каждого пути:
Формируется полный URL, объединяя address и way.
Выполняется GET-запрос по этому URL.
Ответ сервера преобразуется в JSON-формат с помощью .json().
Предполагается, что JSON-ответ представляет собой список чисел.
Функция sum() суммирует все числа из этого списка.
Полученная сумма добавляется к summ.

Принцип работы:
Пользователь вводит адрес сервера.
Пользователь вводит список путей (по одному пути на строку) через стандартный ввод.
Программа отправляет GET-запрос по каждому пути, добавляя его к базовому адресу.
Программа суммирует все числа, полученные из каждого JSON-ответа.
Программа выводит итоговую сумму всех чисел.

Список пользователей

```python
from requests import get

address = 'http://' + input() + '/users'
data = get(address).json()
names = []
for i in data:
    names.append(f"{i['last_name']} {i['first_name']}")
for name in sorted(names):
    print(name)
```

Этот код получает список пользователей с сервера и выводит их имена в отсортированном порядке.
1. Импорт библиотеки и ввод адреса:
from requests import get
address = 'http://' + input() + '/users'
Формируется URL адреса, объединяя введенный пользователем адрес и путь /users.
2. Получение данных и преобразование в JSON:
data = get(address).json()
Выполняется GET-запрос по сформированному адресу.
Ответ сервера преобразуется в JSON формат и сохраняется в переменную data.
3. Создание списка имен:
names = []
for i in data:
    names.append(f"{i['last_name']} {i['first_name']}")
Создается пустой список names для хранения имен пользователей.
Цикл for перебирает каждый элемент i в списке data. Предполагается, что data - это список словарей, где каждый словарь содержит информацию о пользователе.
Для каждого пользователя формируется строка с именем, объединяя фамилию (i['last_name']) и имя (i['first_name']).
Сформированная строка добавляется в список names.
4. Сортировка и вывод имен:
for name in sorted(names):
    print(name)
Функция sorted(names) сортирует список имен по алфавиту.
Цикл for перебирает каждое имя name в отсортированном списке и выводит его на консоль.
Принцип работы:
Получение данных: Программа получает список пользователей в формате JSON с сервера по указанному адресу.
Извлечение имен: Программа извлекает фамилию и имя каждого пользователя и формирует строку с полным именем.
Сортировка: Программа сортирует список имен по алфавиту.


Рассылка сообщений

```python
from requests import get
from sys import stdin

address = 'http://' + input() + '/users/' + input()
d = ''.join(i for i in stdin)
data = {}
try:
    data = get(address).json()
except ValueError:
    print('Пользователь не найден')
if data:
    for key in data:
        d = d.replace('{' + key + '}', str(data[key]))
    print(d)  
```


Импорт библиотек:
requests: для выполнения HTTP-запросов.
sys: для чтения данных из стандартного потока ввода (stdin).
Получение адреса:
Спрашивает у пользователя адрес сервера и имя пользователя, формируя полный URL.
Чтение шаблона:
Считывает весь текст из стандартного потока ввода (stdin) и сохраняет его в переменной d. Предполагается, что этот текст содержит шаблон с метками вида {ключ}, которые будут заменены на значения из JSON.
Получение данных с сервера:
Выполняет GET-запрос к сформированному URL.
Пытается распарсить ответ сервера как JSON и сохранить результат в data.
Если возникает ошибка ValueError (например, сервер не вернул JSON или вернул некорректный JSON), выводит сообщение "Пользователь не найден".
Замена меток в шаблоне:
Если данные data получены успешно (не пустой словарь), выполняется цикл по ключам в data.
Для каждого ключа происходит замена всех вхождений {ключ} в шаблоне d на соответствующее значение из data (преобразованное в строку).
Вывод результата:
Выводит на консоль итоговый текст, в котором метки из шаблона заменены на значения из JSON.
Пример использования:
Входные данные (ввод пользователя):
Адрес сервера: example.com
Имя пользователя: John
Шаблон (введён в stdin):
Привет, {name}!
Твой email: {email}.
Ответ сервера (JSON):
{"name": "John Doe", "email": "john.doe@example.com"}

Результат (вывод программы):
Привет, John Doe!
Твой email: john.doe@example.com.

В итоге, этот код:
Позволяет создавать шаблоны текста с метками.
Заполняет эти шаблоны данными, полученными с сервера в формате JSON.

Регистрация нового пользователя

```python
from requests import post
from json import dumps

address = 'http://' + input() + '/users'
data = {}
data["username"] = input()
data["last_name"] = input()
data["first_name"] = input()
data["email"] = input()
post(address, data=dumps(data))
```

Этот код собирает информацию о пользователе и отправляет её на сервер в формате JSON с помощью HTTP POST-запроса.
Импорт библиотек:
requests: для выполнения HTTP-запросов.
json: для работы с JSON-данными.
Получение адреса:
Спрашивает у пользователя адрес сервера и добавляет к нему путь /users, формируя URL для отправки данных.
Сбор данных пользователя:
Создаёт пустой словарь data.
Запрашивает у пользователя значения для следующих ключей:
username
last_name
first_name
email
Сохраняет введённые значения в соответствующие ключи словаря data.
Отправка данных на сервер:
dumps(data): преобразует словарь data в JSON-строку.
post(address, data=...): выполняет POST-запрос к указанному адресу, передавая JSON-данные в теле запроса.
В итоге, этот код:
Запрашивает у пользователя адрес сервера.
Запрашивает у пользователя информацию: имя пользователя, фамилию, имя и email.
Отправляет эту информацию на сервер в формате JSON с помощью POST-запроса.

Изменение данных

```python
from requests import put
from json import dumps
from sys import stdin

address = 'http://' + input() + '/users/' + input()
line = [i.strip().split('=') for i in stdin]
data = {}
for j in line:
    data[j[0]] = j[1]
put(address, data=dumps(data))
```

Этот код обновляет информацию о пользователе на сервере, используя данные, введённые в формате "ключ=значение".
Импорт библиотек:
requests: для выполнения HTTP-запросов.
json: для работы с JSON-данными.
sys: для чтения данных из стандартного потока ввода (stdin).
Получение адреса:
Спрашивает у пользователя адрес сервера и имя пользователя, формируя полный URL для обновления данных.
Чтение данных из stdin:
line = [i.strip().split('=') for i in stdin]:
Читает строки из стандартного потока ввода.
Каждую строку:
Очищает от пробелов в начале и конце (strip()).
Разбивает по знаку = на две части (split('=')).
Результат - список списков, где каждый вложенный список содержит ключ и значение.
Формирование словаря данных:
Создаёт пустой словарь data.
Цикл for j in line: для каждой пары "ключ=значение" из списка line:
Добавляет в словарь data элемент с ключом j[0] (ключ) и значением j[1] (значение).
Отправка данных на сервер:
dumps(data): преобразует словарь data в JSON-строку.
put(address, data=...): выполняет PUT-запрос к указанному адресу, передавая JSON-данные в теле запроса. PUT-запрос используется для обновления существующих ресурсов.
В итоге, этот код:
Запрашивает у пользователя адрес сервера и имя пользователя.
Читает данные из стандартного потока ввода в формате "ключ=значение".
Формирует словарь с этими данными.
Отправляет словарь в формате JSON на сервер с помощью PUT-запроса, чтобы обновить информацию о пользователе.

Удаление данных

```python
from requests import delete

address = 'http://' + input() + '/users/' + input()
delete(address)
```

Импорт библиотеки:
from requests import delete: Импортирует функцию delete из библиотеки requests, которая используется для отправки HTTP DELETE запросов.
Формирование адреса:
address = 'http://' + input() + '/users/' + input():
Создает переменную address для хранения URL адреса.
Сначала добавляется базовый URL http://.
Затем дважды вызывается функция input(), чтобы получить от пользователя:
Часть адреса до /users/: Например, имя хоста или IP-адрес сервера.
Идентификатор пользователя: Например, ID или имя пользователя, который нужно удалить.
Эти части объединяются в один URL адрес.
Отправка DELETE запроса:
delete(address): Отправляет HTTP DELETE запрос на сформированный URL адрес. Это означает запрос на удаление ресурса, связанного с этим адресом (в данном случае, пользователя).
