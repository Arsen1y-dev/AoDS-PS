6. Распечатать файловую структуру текущей директории с учётом вложенности.

```python
def print_directory_structure(directory):
    for root, dirs, files in os.walk(directory):
        level = root.replace(directory, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f"{indent}{os.path.basename(root)}/")
        sub_indent = ' ' * 4 * (level + 1)
        for file in files:
            print(f"{sub_indent}{file}")

current_directory = os.getcwd()
print("Файловая структура текущей директории:")
print_directory_structure(current_directory)
```

10. Проверить, успешно ли создан файл и является ли он файлом или папкой.

```python
file_name = input("Введите имя файла: ")
if os.path.exists(file_name):
    if os.path.isfile(file_name):
        print(f"'{file_name}' существует и является файлом.")
    elif os.path.isdir(file_name):
        print(f"'{file_name}' существует и является папкой.")
else:
    print(f"'{file_name}' не существует.")
```

11. Распечатать абсолютный путь до созданного файла и его размер.

```python
file_name = input("Введите имя файла: ")
if os.path.exists(file_name):
    abs_path = os.path.abspath(file_name)
    size = os.path.getsize(file_name)
    print(f"Абсолютный путь до файла '{file_name}': {abs_path}")
    print(f"Размер файла '{file_name}': {size} байт")
else:
    print(f"'{file_name}' не существует.")
```

12. Скопировать файл на уровень выше под тем же именем.

```python
file_name = input("Введите имя файла: ")
if os.path.exists(file_name):
    shutil.copy(file_name, os.path.join(os.path.dirname(os.getcwd()), file_name))
    print(f"Файл '{file_name}' успешно скопирован на уровень выше.")
else:
    print(f"'{file_name}' не существует.")
```

```python
import shutil

folder_name = 'test'
if os.path.exists(folder_name):
    shutil.rmtree(folder_name)
    print(f"Папка '{folder_name}' успешно удалена со всем содержимым.")
else:
    os.chdir('..')
    shutil.rmtree(folder_name)
```
